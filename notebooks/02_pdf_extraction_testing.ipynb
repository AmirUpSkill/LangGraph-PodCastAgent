````python
import sys 
import os 
import time 
import pathlib 
import textwrap
from IPython.display import Markdown as md , display 
print(f"Python {sys.version.split()[0]}  |  working dir: {pathlib.Path.cwd()}")

````

````python
# --- Helper for pretty timing & display --- 
def timed(fn, *a, **kw):
    t0 = time.perf_counter()
    out = fn(*a, **kw)
    return out, f"{(time.perf_counter()-t0)*1000:.0f} ms"
````

````python
# --- Load Test PDF --- 
pdf_path = pathlib.Path("Test.pdf")
assert pdf_path.exists(), "ðŸ“„  Test.pdf must live in the same folder as this notebook"
display(md(f"**Input file**: `{pdf_path}`  ({pdf_path.stat().st_size/1024:.1f} KB)"))
````

````python
# --- #1 : Test it with PyPDF --- 
import pypdf 

def pypdf_text(p: pathlib.Path) -> str:
    txt = ""
    with p.open("rb") as fh:
        reader = pypdf.PdfReader(fh)
        for page in reader.pages:
            txt += page.extract_text()
    return txt
py_txt, py_time = timed(pypdf_text, pdf_path)
display(md(f"### PyPDF  â€“  {py_time}"))
display(md(f"```text\n{textwrap.shorten(py_txt, 600)}...\n```"))
````

````python
import fitz 

def pymupdf_text(p: pathlib.Path) -> str:
    doc = fitz.open(p)
    txt = "\n".join([page.get_text() for page in doc])
    doc.close()
    return txt 
pm_txt, pm_time = timed(pymupdf_text, pdf_path)
display(md(f"### PyMuPDF  â€“  {pm_time}"))
display(md(f"```text\n{textwrap.shorten(pm_txt, 600)}...\n```"))
````

````python
import pdfplumber

def plumber_text(p: pathlib.Path) -> str:
    txt = ""
    with pdfplumber.open(p) as pdf:
        for page in pdf.pages:
            txt += page.extract_text() or ""
    return txt
pl_txt, pl_time = timed(plumber_text, pdf_path)
display(md(f"### pdfplumber  â€“  {pl_time}"))
display(md(f"```text\n{textwrap.shorten(pl_txt, 600)}...\n```"))
````
