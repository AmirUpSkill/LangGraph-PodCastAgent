```python
import sys 
import firecrawl 
from firecrawl import Firecrawl
import json 
import os 
from dotenv import load_dotenv
# --- Load Env --- 
load_dotenv(dotenv_path=os.path.join(os.getcwd(), '.env'))
```

```python
# --- Check Versions --- 
print(f"Python Version : {sys.version}")
print(f"FireCrawl SDK Version : {firecrawl.__version__}")
```

```python
# --- Initialize FireCrawl Client --- 
FIRECRAWL_API_KEY = os.getenv("FIRECRAWL_API_KEY")

if not FIRECRAWL_API_KEY:
    print("ðŸ”¥ ERROR: FIRECRAWL_API_KEY not found in .env file.")
    print("Please ensure your 'notebooks/.env' file is set up correctly.")
else:
    print("âœ… FIRECRAWL_API_KEY loaded successfully.")

# --- Initialize the FireCrawl Client ---
try:
    firecrawl_client = Firecrawl(api_key=FIRECRAWL_API_KEY)
    print("âœ… FireCrawl client initialized successfully.")
except Exception as e:
    print(f"ðŸ”¥ ERROR: Failed to initialize FireCrawl client: {e}")
```

```python
# --- Scrap to Markdown --- 
def scrape_url_to_markdown(url , formats=["markdown","html"]):
    """
        Scrape a URL and return it's Content in Markdown Format.
    """
    try:
        print(f"ðŸš€ Scraping URL: {url}")
        result = firecrawl_client.scrape(
            url=url,
            formats=formats,
            only_main_content=True,
            timeout=30000
        )
        print(f"âœ… Successfully scraped URL: {url}")
        return result
    except Exception as e:
        print(f"ðŸ”¥ ERROR: Failed to scrape URL {url}: {e}")
        return None
```

```python
# --- Quick Test --- 
test_url = "https://blog.langchain.dev/langgraph-cloud/"
result = scrape_url_to_markdown(test_url)

if result:
    print("\nðŸ“‹ Metadata extracted:")
    print(f"Title: {getattr(result.metadata, 'title', 'N/A')}")
    print(f"Description: {getattr(result.metadata, 'description', 'N/A')}")
    print(f"Language: {getattr(result.metadata, 'language', 'N/A')}")
    print(f"Status Code: {getattr(result.metadata, 'statusCode', 'N/A')}")
    
    # Display first 500 characters of markdown
    if hasattr(result, 'markdown'):
        print("\nðŸ“ First 500 characters of markdown content:")
        print(result.markdown[:500] + "..." if len(result.markdown) > 500 else result.markdown)
```
